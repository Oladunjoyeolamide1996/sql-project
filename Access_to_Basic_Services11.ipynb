```python
--Identify regions in Africa
Construct a query that selects only the regions falling within Africa. Use the LIKE operator to identify records where the Region contains Africa.
```

```python
%%sql
SELECT DISTINCT
    Region
FROM
    united_nations.Access_to_Basic_Services
WHERE
    Region LIKE '%Africa%';
```

```python
--Classify SADC countries
Use a CASE statement to classify whether an African country belongs to SADC, or does not belong to SADC. In other words, if the country_name is one of the names we specified as an SADC country (in the table above), it should be classified as SADC. Otherwise, it must be classified as Not Classified.
```

```python
%%sql
SELECT
    Country_name,
    Region,
    CASE
        WHEN Country_name IN (
            'Angola', 'Botswana', 'Democratic Republic of the Congo', 'Comoros',
            'Eswatini', 'Lesotho', 'Madagascar', 'Malawi', 'Mauritius',
            'Mozambique', 'Namibia', 'South Africa', 'United Republic of Tanzania',
            'Zambia', 'Zimbabwe', 'Seychelles'
        ) THEN 'SADC'
        ELSE 'Not Classified'
    END AS SADC_Membership
FROM
    united_nations.Access_to_Basic_Services
WHERE
    Region LIKE '%Africa%';
```

```python
%%sql  
SELECT
    Country_name,
    Region,
    CASE
        -- 🌍 SADC (Southern African Development Community)
        WHEN Country_name IN (
            'Angola', 'Botswana', 'Democratic Republic of the Congo', 'Comoros',
            'Eswatini', 'Lesotho', 'Madagascar', 'Malawi', 'Mauritius',
            'Mozambique', 'Namibia', 'South Africa', 'United Republic of Tanzania',
            'Zambia', 'Zimbabwe', 'Seychelles'
        ) THEN 'SADC'

        -- 🌍 UMA (Union du Maghreb Arabe / Arab Maghreb Union)
        WHEN Country_name IN (
            'Algeria', 'Libya', 'Mauritania', 'Morocco', 'Tunisia'
        ) THEN 'UMA'

        -- 🌍 ECOWAS (Economic Community of West African States)
        WHEN Country_name IN (
            'Benin', 'Burkina Faso', 'Cabo Verde', 'Côte d''Ivoire', 'Gambia',
            'Ghana', 'Guinea', 'Guinea-Bissau', 'Liberia', 'Mali',
            'Niger', 'Nigeria', 'Senegal', 'Sierra Leone', 'Togo'
        ) THEN 'ECOWAS'

        -- ❌ Any other African country not in these groups
        ELSE 'Not Classified'
    END AS REC_Classification
FROM
    united_nations.Access_to_Basic_Services
WHERE
    Region LIKE '%Africa%';
```

```python
--Calculate the minimum, average, and maximum percentages of managed drinking water services for each REC.
Use the MIN(), AVG(), and MAX() aggregate functions on the Pct_managed_drinking_water_services column to obtain the minimum, average, and maximum percentages of managed drinking water services for each REC
Make sure to group the results by the REC using the GROUP BY clause to specify how the data should be grouped for the aggregate calculations.
```

```python
%%sql
SELECT
    CASE
        -- 🌍 SADC (Southern African Development Community)
        WHEN Country_name IN (
            'Angola', 'Botswana', 'Democratic Republic of the Congo', 'Comoros',
            'Eswatini', 'Lesotho', 'Madagascar', 'Malawi', 'Mauritius',
            'Mozambique', 'Namibia', 'South Africa', 'United Republic of Tanzania',
            'Zambia', 'Zimbabwe', 'Seychelles'
        ) THEN 'SADC'

        -- 🌍 UMA (Union du Maghreb Arabe)
        WHEN Country_name IN (
            'Algeria', 'Libya', 'Mauritania', 'Morocco', 'Tunisia'
        ) THEN 'UMA'

        -- 🌍 ECOWAS (Economic Community of West African States)
        WHEN Country_name IN (
            'Benin', 'Burkina Faso', 'Cabo Verde', 'Côte d''Ivoire', 'Gambia',
            'Ghana', 'Guinea', 'Guinea-Bissau', 'Liberia', 'Mali',
            'Niger', 'Nigeria', 'Senegal', 'Sierra Leone', 'Togo'
        ) THEN 'ECOWAS'

        ELSE 'Not Classified'
    END AS REC_Classification,

    -- 🧮 Aggregate functions
    MIN(Pct_managed_drinking_water_services) AS Min_Drinking_Water_Pct,
    AVG(Pct_managed_drinking_water_services) AS Avg_Drinking_Water_Pct,
    MAX(Pct_managed_drinking_water_services) AS Max_Drinking_Water_Pct

FROM
    united_nations.Access_to_Basic_Services
WHERE
    Region LIKE '%Africa%'
GROUP BY
    REC_Classification;
```
